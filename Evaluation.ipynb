```python
from search_engine import SearchEngine

s = SearchEngine()
```

```python
from search_modules import Query
import numpy as np



def sh(d):
    for idx, doc in d.iterrows():
        print('#',idx) # , ' || ', doc.score, ' || PR=',doc.pr_score,' || COS=',doc.cos_score)
        print(' '.join(doc.text.split()[0:500]))
        print('')
        
        
def cand_docs_query(q):
    a = np.arange(4) / 3
    idx = set()
    for alpha in a:
        x = s.run_query(Query(q),alpha_pr=alpha)
        x = x[0:10]
        for i in x.index:
            idx.add(i)
    print(idx)
    return list(idx)
    
cand_NASA = cand_docs_query('NASA Apollo')
cand_SOUND = cand_docs_query('sound driver')
cand_PRINTER = cand_docs_query('color printer')
```

```python
d = s.newsgroup_frame
```

```python
d_NASA = d.iloc[cand_NASA]
sh(d_NASA)
rel_NASA = set([9350,10658,7208,4649,6490])
```

```python
d_SOUND = d.iloc[cand_SOUND]
sh(d_SOUND)
rel_SOUND = set([3167,9563,10812,8094,5767,9754,10116,2946])
```

```python
d_PRINTER = d.iloc[cand_PRINTER]
sh(d_PRINTER)
rel_PRINTER = set([4236,10473])
```

```python
a2 = a
a = np.arange(10)/10

def p_at_r(rel_ids, df):
    k = len(rel_ids)
    idx = df.index.tolist()
    f = 0
    for i in range(k):
        docID = idx[i]
        if docID in rel_ids:
            f += 1
    pr = f / k
    return pr
  
print('NASA apollo')
for alpha in a:
    pr = p_at_r(rel_NASA,s.run_query(Query('NASA apollo'),alpha_pr=alpha))
    print('r-precision, a=',alpha,' -> ', pr)
    
print()
print('sound driver')
for alpha in a:
    pr = p_at_r(rel_SOUND,s.run_query(Query('sound driver'),alpha_pr=alpha))
    print('r-precision, a=',alpha,' -> ', pr)

print()
print('color printer')
for alpha in a:
    pr = p_at_r(rel_PRINTER,s.run_query(Query('color printer'),alpha_pr=alpha))
    print('r-precision, a=',alpha,' -> ', pr)
```
